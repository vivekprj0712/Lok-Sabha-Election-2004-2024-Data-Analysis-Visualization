#Analyzing Voter Turnout Trends

```python
print(df[['State', 'Turnout', 'Year']].isnull().sum())
```

```python
# Plot the voter turnout percentage for each state across different years
sns.lineplot(data=state_turnout_by_year, x='Year', y='Turnout', hue='State', marker='o')

# Add plot title and labels
plt.title('Voter Turnout Percentage by State and Year (2004-2024)', fontsize=16)
plt.xticks([2004, 2009, 2014, 2019, 2024])
plt.xlabel('Year', fontsize=14)
plt.ylabel('Voter Turnout Percentage', fontsize=14)
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
```

#Party Performance Evaluation

```python
#Data Check
print(df[['Party', 'Year']].isnull().sum())
```

```python
#Calculate the Number of Seats Won by Each Party
seats_won = df.groupby(['Year', 'Party']).size().reset_index(name='Seats')
seats_won
```

```python
#Calculate the Total Number of Seats in Each Year:
total_seats = df.groupby('Year').size().reset_index(name='Total_Seats')
total_seats
```

```python
party_performance = pd.merge(seats_won, total_seats, on='Year')
party_performance['Percentage_Seats'] = (party_performance['Seats'] / party_performance['Total_Seats']) * 100
```

```python
party_performance
```

```python
# Set plot style
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=party_performance, x='Year', y='Percentage_Seats', hue='Party', marker='o')
# Add plot title and labels
plt.title('Party Performance Trend (2004-2024)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Percentage of Seats Won (%)', fontsize=14)
plt.legend(title='Party', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```
